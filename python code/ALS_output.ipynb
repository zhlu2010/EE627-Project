```python
fpredict = open('predictScoreRank15.txt','r')
```

```python
predictDic = {}
for line in fpredict:
    record = line.strip().split('|')
    user = record[0]
    item = record[1]
    score = float(record[2])
    predictDic.setdefault(user, {})[item] = score
predictDic
```

```python
for key in predictDic:
    predictDic[key] = sorted(predictDic[key].items(), key = lambda x : x[1], reverse = True)
predictDic
```

```python
####
predictDic['199810']
```

```python
item_score = []
for key in predictDic:
    for i in range(6):
        if i < 3:
            need1 = key + '_' + predictDic[key][i][0] + '|' + str(1)
            item_score.append(need1)
        else:
            need2 = key + '_' + predictDic[key][i][0] + '|' + str(0)
            item_score.append(need2)
```

```python
import pandas as pd
```

```python
result_list = []
for item in item_score:
    line = item.strip().split('|')
    ID = line[0]
    score = line[1]
    result = [ID, score]
    result_list.append(result)
name = ['TrackID', 'Predictor']
test=pd.DataFrame(columns=name, data=result_list)
test.to_csv('ALS_rank15.csv',encoding='gbk')
```
