```python
import numpy
dataDir='/Users/kunpenghao/Desktop/Study_in_SIT/EE_627/Project/all/'
testfile = dataDir + 'testItem2.txt'
trainfile = dataDir + 'trainItem2.txt'
trackfile = dataDir + 'trackData2.txt'
output_file= dataDir + 'output1.txt'

ftest = open(testfile,'r')
fTrain=open(trainfile, 'r')
fTrack = open(trackfile, 'r')
#Trainline= fTrain.readline()
fOut = open(output_file, 'w')
```

```python
#read testItem2.    username as KEY, item as VALUE
userDic = {}
for line in ftest:
    if '|' in line:
        ID=line.strip().split('|')
        userID = ID[0]
        key = userID
    else:
        value = line.strip('\n')
        userDic.setdefault(key,[]).append(value)
```

```python
#combine test list as username|item.
testList = []
for key in userDic:
    for value in userDic[key]:
        outtext = str(key) + '|' + str(value)
        testList.append(outtext)
```

```python
####
testList
```

```python
#read trainItem2 as list like   username|item|score

trainList = []
for line in fTrain:
    if '|' not in line:
        continue
    else:
        nameline = line.strip().split('|')
        userN = nameline[0]
        num = nameline[1]
        num = int(num)
        for i in range(num):
            scoreline = next(fTrain).split()
            scoreID = scoreline[0]
            score = scoreline[1]
            text = str(userN) + '|' + str(scoreID) + '|' + str(score)
            trainList.append(text)
```

```python
####
trainList
```

```python
# create .txt file for test data
test_txt=[]
for record in testList:
    record = record + '|0' + '\n'
    test_txt.append(record)
f = open('testList.txt', 'a')
f.writelines(test_txt)
f.close()
```

```python
# create .txt file for train data
train_txt = []
for record in trainList:
    record = record + '\n'
    train_txt.append(record)
f = open('trainList.txt', 'a')
f.writelines(train_txt)
f.close()
```
